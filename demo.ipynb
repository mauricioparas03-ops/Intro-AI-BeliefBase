# Belief Revision Agent Demo
This notebook demonstrates the core functionality of our belief revision engine, including logical entailment and partial meet contraction.

```python
# Import our core modules
from knowledge_representation import BeliefBase, Atom
from inference_engine import entails
from revision_engine import contract

# Initialize Atoms
p = Atom('p')
q = Atom('q')
r = Atom('r')
s = Atom('s')
```

### 1. Testing Logical Entailment
We initialize a knowledge base with `{~p -> q, q -> p, p -> (r & s)}` and test if it logically entails `p & r & s`.

```python
kb = BeliefBase()
kb.add(~p >> q)
kb.add(q >> p)
kb.add(p >> (r & s))

test_formula = p & r & s
print(f"Does KB entail '{test_formula}'?")
print(f"Answer: {entails(kb, test_formula)}")
```
